# Преобразование данных

**1.** Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def get_education_level(education_string):
    """Функция для получения уровня образования

    Args:
        education_string (sting): Строка с данными по образованию соискателя.

    Returns:
        string: Строка для установки категории
    """
    word_list = education_string.split(' ')
    fist_two_words = word_list[0] + ' ' + word_list[1]
    if fist_two_words == 'Высшее образование':
        return 'высшее'
    elif fist_two_words == 'Неоконченное высшее':
        return 'неоконченное высшее'
    elif fist_two_words == 'Среднее специальное':
        return 'среднее специальное'
    elif fist_two_words == 'Среднее образование':
        return 'среднее'
    else:
        return 'другое'
    

df['Образование'] = df['Образование и ВУЗ'].apply(get_education_level)
df['Образование'] = df['Образование'].astype('category')
print(df['Образование'].unique())

mask = df['Образование'] == 'среднее'
print('Количество соискателей со средним образованием:', df[mask]['Образование'].shape[0])

df.drop(['Образование и ВУЗ'], axis=1, inplace=True)
```

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '.

```python
# Выделим признак пола по первой букве из признака "Пол, возраст"
df['Пол'] = df['Пол, возраст'].apply(lambda x: x[:1])

mask = df['Пол'] == 'Ж'
print('Процент женских резюме:', round(df[mask]['Пол'].shape[0] / df['Пол'].shape[0] * 100, 2))

def get_age(gender_age_string):
    """Функция выделения возраста

    Args:
        gender_age_string (sting): Строка с данными по полу и возрасту.

    Returns:
        int: Возраст соискателя 
    """    
    part_list = gender_age_string.split(' , ')
    age_list = part_list[1].split(' ')
    return int(age_list[1])


df['Возраст'] = df['Пол, возраст'].apply(get_age)

print('Средний возраст соискателей:', round(df['Возраст'].mean(), 1))

df.drop(['Пол, возраст'], axis=1, inplace=True)
```

**3.** Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-третьих, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_work_months(experience_string):
    """Функция для получения количества месяцев опыта работы

    Args:
        experience_string (sting): Строка с данными по опыту работы соискателя (годы, месяцы).

    Returns:
        int: Количество месяцев опыта работы соискателя
    """
    if pd.isna(experience_string):
        return experience_string
    
    part1 = experience_string.split(',')[0]
    if 'Опыт работы: ' in part1:
        part2 = part1[13:]
    elif 'Опыт работы ' in part1:
        part2 = part1[12:]
    else:
        return 0   # Не найдена точка отсчета опыта работы - возвращаем ноль!
    split3 = part2.split(' ')
    
    years = 0   # годы опыта работы
    months = 0   # месяцы года работы
    
    # Разбор строки с числовыми значениями опыта работы (годы, месяцы)
    year_list = ['год', 'года', 'лет']
    months_list = ['месяц', 'месяца', 'месяцев']
    if len(split3) > 2 and split3[1] in year_list and split3[3] in months_list:
        years = '0' if len(split3[0]) == 0 else split3[0]
        months = '0' if len(split3[2]) == 0 else split3[2]
    elif split3[1] in year_list:
        years = '0' if len(split3[0]) == 0 else split3[0]
        months = 0
    elif split3[1] in months_list:
        years = 0
        months = '0' if len(split3[0]) == 0 else split3[0]
    
    return int(years) * 12 + int(months)


df['Опыт работы (месяц)'] = df['Опыт работы'].apply(get_work_months)

df.drop(['Опыт работы'], axis=1, inplace=True)

print('Медианный опыт работы (в месяцах):', df['Опыт работы (месяц)'].median())
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city(complex_string):
    """Функция получения города, категории города

    Args:
        complex_string (sting): Строка с данными по городу, перездам и компндировкам

    Returns:
        string: Строка с названием города или категории города 
    """
    city_string = complex_string.split(' , ')[0]
    
    capital_cities = ['Москва', 'Санкт-Петербург']
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    
    if city_string in capital_cities:
        return city_string
    elif city_string in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'


df['Город'] = df['Город, переезд, командировки'].apply(get_city)
df['Город'] = df['Город'].astype('category')

mask = df['Город'] == 'Санкт-Петербург'
print('% соискателей живущих в Санкт-Петербурге:', round(df[mask]['Город'].shape[0] / df['Город'].shape[0] * 100, 0))
```

```python
def get_binary_ready(complex_string, mode):
    """Функция получения бинарной готовности к переезду и к командировке

    Args:
        complex_string (sting): Строка с данными по городу, перездам и компндировкам
        mode (string): Режим работы функции ('переезд' или 'командировка')

    Returns:
        boolean: Истина - готов к перезду/командировкам, ложь - не готов к перезду/командировкам 
    """
    int_mode = 2 if mode == 'переезд' else 3

    comlex_split = complex_string.split(' , ')
    ready_string = ''
    if len(comlex_split) == 4:
        ready_string = comlex_split[int_mode]
    elif len(comlex_split) == 3:
        ready_string = comlex_split[int_mode-1]
   
    return not('не ' in ready_string)


df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(get_binary_ready, mode='переезд')
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(get_binary_ready, mode='командировка')

df.drop(['Город, переезд, командировки'], axis=1, inplace=True)

mask = df['Готовность к переезду'] & df['Готовность к командировкам']
print('% соискателей готовых к переезду и к командировкам:', round(df[mask].shape[0] / df.shape[0] * 100, 0))
```

**5.** Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Обработка признака занятости
list_busyness = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
for item in list_busyness:
    df[item] = df['Занятость'].apply(lambda x: item in x)

mask = df['проектная работа'] & df['волонтерство']
print('Проектную работу или волонтерство ищут:', df[mask].shape[0])

# Обработка признака "График"
list_schedule = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
for item in list_schedule:
    df[item] = df['График'].apply(lambda x: item in x)

mask = df['вахтовый метод'] & df['гибкий график']
print('Работу вахтовым методом или с гибким графиком ищут:', df[mask].shape[0])

df.drop(['Занятость', 'График'], axis=1, inplace=True)
```

2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
ExchangeRates = pd.read_csv('../../data/ExchangeRates.csv', sep=',')
ExchangeRates.drop(['per', 'time', 'vol'], axis=1, inplace=True)
ExchangeRates['date'] = pd.to_datetime(ExchangeRates['date'])

df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'])
df['date'] = pd.to_datetime(df['Обновление резюме'].dt.date)

# Выделение суммы ЗП и валюты в отдельные признаки для дальнейшей обработки
df['ЗП_исх'] = df['ЗП'].apply(lambda x: x.split(' ')[0])
df['ЗП_исх'] = df['ЗП_исх'].astype('float64')
df['currency'] = df['ЗП'].apply(lambda x: x.split(' ')[1])

def currency_name(in_base_name):
    """Функция преобразования наименования валюты по ISO

    Args:
        in_base_name (sting): Строка с названием валюты в базе

    Returns:
        string: Строка с наименованеим валюты по стандарту ISO
    """
    # Будем обрабатывать только не сопадающие по именам валюты
    change_dic = {'бел.руб.':'BYN', 'грн.':'UAH', 'сум':'UZS'}
    try:
        result = change_dic[in_base_name]
    except KeyError:
        result = in_base_name        
    return result


df['currency'] = df['currency'].apply(currency_name)

# Присоединение к исходным данным таблицы курсов соответствующей валюты на нужную дату 
df = df.merge(ExchangeRates, on=['currency','date'], how='left')

# Заполнение единицей (1) отсутствующих данных курса и кратности для базовой валюты (руб.)
df['close'] = df.apply(lambda x: 1 if x['currency'] == 'руб.' else x['close'], axis=1)
df['proportion'] = df.apply(lambda x: 1 if x['currency'] == 'руб.' else x['proportion'], axis=1)

# Пересчет ЗП в рубли
df['ЗП (руб)'] = df['ЗП_исх'] * df['close'] / df['proportion']

print('Медианная заработная плата соискателей (в тыс. руб.):', round(df['ЗП (руб)'].median()/1000, 0))

df.drop(['ЗП', 'ЗП_исх', 'Обновление резюме'], axis=1, inplace=True)
```
